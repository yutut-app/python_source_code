# Pythonのデータ型入門

## はじめに
こんにちは！これからPythonプログラミングを学ぶ皆さんに、Pythonの基本的なデータ型について説明します。データ型とは、プログラムの中でデータを表現するための種類のことです。Pythonには、int型、float型、str型、bool型などがあります。これらのデータ型を理解することは、Pythonでプログラムを書く上で非常に重要です。

## 1. int型(Integers,整数型)
int型は、正の整数、0、負の整数を表すデータ型です。

```python
# 変数ageにint型の値25を代入
age = 25
# type()関数で変数ageのデータ型を確認
print(type(age))
```

上記の例では、ageという変数に25という整数値を代入しています。type()関数を使うと、変数のデータ型を確認できます。この場合、ageはint型であることがわかります。

## 2.float型(Floats,浮動小数点数型)
float型は、小数点を含む実数を表現するためのデータ型です。

```python
# 変数heightにfloat型の値175.5を代入
height = 175.5
# type()関数で変数heightのデータ型を確認
print(type(height))
```

上記の例では、heightという変数に175.5という小数値を代入しています。type()関数で確認すると、heightはfloat型であることがわかります。

# 3. str型(Strings,文字列型)

str型は、文字の並びを表現するためのデータ型です。文字列は、シングルクォート（'）またはダブルクォート（"）で囲みます。

```python
# 変数nameにstr型の値"太郎"を代入
name = "太郎"
# type()関数で変数nameのデータ型を確認
print(type(name))
```

上記の例では、nameという変数に"太郎"という文字列を代入しています。type()関数で確認すると、nameはstr型であることがわかります。

# 4. bool型(Booleans,真偽値型)

bool型は、真（True）または偽（False）のいずれかの値を表現するためのデータ型です。

```python
# 変数is_studentにbool型の値Trueを代入
is_student = True
# type()関数で変数is_studentのデータ型を確認
print(type(is_student))
```

上記の例では、is_studentという変数にTrueという真偽値を代入しています。type()関数で確認すると、is_studentはbool型であることがわかります。

# 5. 型変換
型変換とは、あるデータ型から別のデータ型に変換することです。Pythonでは、int()、float()、str()などの関数を使って型変換を行います。

```python
# int型からfloat型への変換
age = 25
age_float = float(age)
print(age_float)

# float型からint型への変換（小数点以下の情報は失われる）
height = 175.5
height_int = int(height)
print(height_int)

# int型からstr型への変換
score = 80
score_str = str(score)
print(score_str)
```

上記の例では、ageというint型の変数をfloat()関数でfloat型に変換しています。同様に、heightというfloat型の変数をint()関数でint型に変換しています。ただし、float型からint型への変換では、小数点以下の情報が失われることに注意が必要です。また、scoreというint型の変数をstr()関数でstr型に変換しています。
