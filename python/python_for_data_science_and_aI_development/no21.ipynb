# Pythonで箱ひげ図を作ろう！日本の人気観光地データを可視化

## はじめに

こんにちは！このページでは、Pythonを使って箱ひげ図（"Box Plot"）を作成する方法を学びます。箱ひげ図は、データの分布を視覚的に表現するのに適した図で、観光客数の推移など、様々なデータを分析するのに役立ちます。日本の人気観光地データを例に、箱ひげ図の作り方を step by step で説明していきます。

## 箱ひげ図とは

箱ひげ図は、データの分布を5つの要素で表現します：
1. 最小値：外れ値を除いた最小値
2. 第1四分位数：データの25%がこの値以下
3. 中央値：データの50%がこの値以下
4. 第3四分位数：データの75%がこの値以下
5. 最大値：外れ値を除いた最大値

また、箱ひげ図は外れ値も点で表示します。

## データの準備

まず、分析に使うデータを準備します。


````python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの作成（2013年から2024年までの12年間のデータ）
data = {
    '年': range(2013, 2025),
    '東京': [1000, 1200, 1400, 1600, 1800, 1000, 1300, 1500, 1700, 1900,0,200],
    '京都': [800, 900, 1000, 1100, 1200, 700, 900, 1000, 1100, 1300,1900,2100],
    '大阪': [900, 1000, 1100, 1300, 1400, 800, 1000, 1200, 1300, 1500,2100,2400],
    '北海道': [700, 800, 900, 1000, 1100, 600, 800, 900, 1000, 1200,1800,2200]
}

df = pd.DataFrame(data)
df.set_index('年', inplace=True)

print(df.head())
````

このコードでは、4つの観光地（東京、京都、大阪、北海道）の12年間の観光客数データを作成しています。`set_index('年', inplace=True)`で、'年'列をインデックスに設定しています。

## 基本的な箱ひげ図の作成

次に、基本的な箱ひげ図を作成します。


````python
# 基本的な箱ひげ図の作成
df.plot(kind='box', figsize=(10, 6))
plt.title('日本の主要観光地の年間観光客数分布（2013-2024）')
plt.ylabel('観光客数（万人）')
plt.xlabel('観光地')

# x軸のラベルを設定
plt.xticks(range(1, len(df.columns) + 1), df.columns)

plt.show()
````

ここでは、`df.plot()`メソッドを使用し、`kind='box'`で箱ひげ図を指定しています。`figsize=(10, 6)`でグラフのサイズを設定しています。

## 箱ひげ図の解釈

箱ひげ図から以下の情報を読み取ることができます：
- 箱の下端は第1四分位数、上端は第3四分位数を示します。
- 箱の中の線は中央値を示します。
- ひげの端は、箱の長さの1.5倍以内にあるデータの最小値と最大値を示します。
- 点で表示されているのは外れ値です。

## 水平方向の箱ひげ図

箱ひげ図は水平方向に表示することもできます。


````python
# 水平方向の箱ひげ図
df.plot(kind='box', figsize=(10, 6), vert=False)
plt.title('日本の主要観光地の年間観光客数分布（2013-2024）')
plt.xlabel('観光客数（万人）')
plt.ylabel('観光地')

# y軸のラベルを設定
plt.yticks(range(1, len(df.columns) + 1), df.columns)

plt.show()
````

`vert=False`パラメータを追加することで、箱ひげ図を水平方向に表示できます。

## 詳細な統計情報の表示

箱ひげ図に表示されている統計情報を数値で確認するには、`describe()`メソッドを使用します。


````python
# 詳細な統計情報の表示
print(df.describe())
````

この結果から、各観光地の観光客数の最小値、第1四分位数、中央値、第3四分位数、最大値などを確認できます。

問題1
あなたは、過去5年間の4つの異なる科目（国語、数学、英語、理科）の試験結果を記録しました。このデータを使用して、各科目の成績分布を表す箱ひげ図を作成してください。

データ：
```python
data = {
    '年度': [2019, 2020, 2021, 2022, 2023],
    '国語': [75, 82, 78, 85, 80],
    '数学': [68, 72, 80, 75, 78],
    '英語': [70, 75, 82, 78, 85],
    '理科': [72, 78, 75, 80, 76]
}
```

以下の要件を満たす箱ひげ図を作成してください：
1. グラフのタイトルを「過去5年間の科目別成績分布」とする
2. y軸のラベルを「点数」とする
3. x軸のラベルを各科目名にする
4. グラフのサイズを(10, 6)に設定する
5. 詳細な統計情報も表示する

解答例(Pythonコード):


````python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの準備
data = {
    '年度': [2019, 2020, 2021, 2022, 2023],
    '国語': [75, 82, 78, 85, 80],
    '数学': [68, 72, 80, 75, 78],
    '英語': [70, 75, 82, 78, 85],
    '理科': [72, 78, 75, 80, 76]
}

df = pd.DataFrame(data)
df.set_index('年度', inplace=True)

# データの確認
print(df)

# 箱ひげ図の作成
df.plot(kind='box', figsize=(10, 6))
plt.title('過去5年間の科目別成績分布')
plt.ylabel('点数')
plt.xlabel('科目')

# x軸のラベルを設定
plt.xticks(range(1, len(df.columns) + 1), df.columns)

plt.show()

# 詳細な統計情報の表示
print("\n詳細な統計情報:")
print(df.describe())
````

解説:
1. 必要なライブラリ（pandas と matplotlib.pyplot）をインポートし、日本語フォントを設定します。
2. 提供されたデータを辞書形式で定義し、それをpandasのDataFrameに変換します。
3. `set_index('年度', inplace=True)` で「年度」列をインデックスに設定します。
4. `print(df)` でデータを確認します。
5. `df.plot(kind='box', figsize=(10, 6))` で箱ひげ図を作成します。`kind='box'`で箱ひげ図を指定し、`figsize=(10, 6)`でグラフのサイズを設定します。
6. `plt.title()`, `plt.ylabel()`, `plt.xlabel()` でグラフのタイトルと軸ラベルを設定します。
7. `plt.xticks()` でx軸のラベルを各科目名に設定します。
8. `plt.show()` でグラフを表示します。
9. `df.describe()` で詳細な統計情報を表示します。

この箱ひげ図から、各科目の成績分布を視覚的に比較できます。中央値、四分位範囲、最小値、最大値などの情報が一目で分かります。また、詳細な統計情報から具体的な数値も確認できます。


問題2
あなたは、5人の友人と6ヶ月間、毎月の運動時間を記録しました。このデータを使用して、各人の運動時間の分布を表す水平方向の箱ひげ図を作成してください。

データ：
```python
data = {
    '月': ['1月', '2月', '3月', '4月', '5月', '6月'],
    'あなた': [30, 45, 0, 40, 50, 55],
    '友人A': [20, 35, 40, 30, 45, 50],
    '友人B': [50, 55, 65, 60, 70, 75],
    '友人C': [25, 30, 35, 40, 45, 50],
    '友人D': [40, 50, 55, 45, 60, 65],
    '友人E': [35, 40, 45, 50, 55, 60]
}
```

以下の要件を満たす水平方向の箱ひげ図を作成してください：
1. グラフのタイトルを「6ヶ月間の個人別運動時間分布」とする
2. x軸のラベルを「運動時間（分）」とする
3. y軸のラベルを各人の名前にする
4. グラフのサイズを(12, 8)に設定する
5. 詳細な統計情報も表示する

解答例(Pythonコード):


````python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの準備
data = {
    '月': ['1月', '2月', '3月', '4月', '5月', '6月'],
    'あなた': [30, 45, 60, 40, 50, 120],
    '友人A': [20, 35, 40, 30, 45, 5],
    '友人B': [50, 55, 65, 60, 70, 150],
    '友人C': [25, 30, 35, 40, 45, 50],
    '友人D': [40, 50, 55, 45, 60, 100],
    '友人E': [35, 40, 45, 50, 55, 0]
}

df = pd.DataFrame(data)
df.set_index('月', inplace=True)

# データの確認
print(df)

# 水平方向の箱ひげ図の作成
plt.figure(figsize=(12, 8))
df.boxplot(vert=False, meanline=True, showmeans=True)
plt.title('6ヶ月間の個人別運動時間分布')
plt.xlabel('運動時間（分）')
plt.ylabel('名前')

plt.show()

# 詳細な統計情報の表示
print("\n詳細な統計情報:")
print(df.describe())
````

解説:
1. 必要なライブラリ（pandas と matplotlib.pyplot）をインポートし、日本語フォントを設定します。
2. 提供されたデータを辞書形式で定義し、それをpandasのDataFrameに変換します。
3. `set_index('月', inplace=True)` で「月」列をインデックスに設定します。
4. `print(df)` でデータを確認します。
5. `df.plot(kind='box', figsize=(12, 8), vert=False)` で水平方向の箱ひげ図を作成します。`kind='box'`で箱ひげ図を指定し、`vert=False`で水平方向に設定します。`figsize=(12, 8)`でグラフのサイズを設定します。
6. `plt.title()`, `plt.xlabel()`, `plt.ylabel()` でグラフのタイトルと軸ラベルを設定します。
7. `plt.yticks()` でy軸のラベルを各人の名前に設定します。
8. `plt.show()` でグラフを表示します。
9. `df.describe()` で詳細な統計情報を表示します。

この水平方向の箱ひげ図から、各人の6ヶ月間の運動時間の分布を視覚的に比較できます。中央値、四分位範囲、最小値、最大値などの情報が一目で分かります。また、詳細な統計情報から具体的な数値も確認できます。友人Bの運動時間が最も長く、ばらつきも大きいことなどが見て取れます。
