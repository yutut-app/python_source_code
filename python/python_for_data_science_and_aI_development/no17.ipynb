# Pythonで折れ線グラフを作ろう！日本の観光データ可視化

## はじめに

こんにちは！このページでは、Pythonを使って折れ線グラフ（"Line Plot"）を作成する方法を学びます。折れ線グラフは、時間の経過に伴うデータの変化を可視化するのに適しています。ここでは、日本の観光データを例に、折れ線グラフの作り方を step by step で説明していきます。

## 折れ線グラフとは

折れ線グラフは、データポイントを直線で結んだグラフです。以下の特徴があります：

1. 時系列データの傾向を把握するのに適している
2. 複数のデータセットを比較しやすい
3. 連続的な変化を表現できる
4. データの急激な変化や異常値を見つけやすい

観光客数の推移や気温の変化など、時間とともに変化するデータの可視化によく使われます。

## データの準備

まずは、分析に使うデータを準備しましょう。ここでは、東京の月別観光客数のデータを使用します。


```python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの作成
data = {
    '月': list(range(1, 13)),
    '観光客数': [100, 120, 150, 200, 250, 220, 300, 350, 280, 220, 180, 150]
}

df = pd.DataFrame(data)
df.set_index('月', inplace=True)

print(df.head())
```

このコードでは、以下の操作を行っています：
1. `import pandas as pd`と`import matplotlib.pyplot as plt`で必要なライブラリをインポートします。pandasはデータ操作用、matplotlibはグラフ描画用です。
2. `plt.rcParams['font.family']`で日本語フォントを設定します。これにより、グラフ上の日本語が正しく表示されます。
3. `data`辞書で月別観光客数のデータを作成します。実際のデータ収集を模倣しています。
4. `pd.DataFrame(data)`でデータフレームを作成します。これにより、データを表形式で扱えるようになります。
5. `set_index('月', inplace=True)`で'月'列をインデックスに設定します。これにより、月をX軸に使用しやすくなります。
6. `print(df.head())`でデータフレームの最初の5行を表示し、データが正しく読み込まれているか確認します。

## 基本的な折れ線グラフの作成

次に、基本的な折れ線グラフを作成します。


```python
# 基本的な折れ線グラフの作成
df.plot(kind='line', figsize=(10, 6))
plt.title('東京の月別観光客数')
plt.ylabel('観光客数（万人）')
plt.xlabel('月')
plt.show()
```



このコードの詳細説明：
1. `df.plot(kind='line')`で線グラフを指定します。`kind='line'`パラメータにより、折れ線グラフが選択されます。
2. `figsize=(10, 6)`でグラフのサイズを設定します。単位はインチです。
3. `plt.title()`でグラフのタイトルを設定します。
4. `plt.ylabel()`と`plt.xlabel()`で、それぞれY軸とX軸のラベルを設定します。
5. `plt.show()`でグラフを表示します。これがないと、グラフが画面に表示されません。

## グラフのカスタマイズ

グラフをより見やすくカスタマイズしましょう。


```python
# グラフのカスタマイズ
df.plot(kind='line', figsize=(10, 6), color='red', marker='o', linewidth=2, markersize=8)
plt.title('東京の月別観光客数（2023年）', fontsize=16)
plt.ylabel('観光客数（万人）', fontsize=12)
plt.xlabel('月', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

このコードでの追加点と詳細説明：
1. `color='red'`で線の色を赤に設定します。他の色名やRGB値も使用可能です。
2. `marker='o'`でデータポイントにマーカーを追加します。'o'は円形のマーカーを意味します。
3. `linewidth=2`で線の太さを設定します。数値が大きいほど太くなります。
4. `markersize=8`でマーカーのサイズを設定します。数値が大きいほど大きくなります。
5. `fontsize`パラメータでフォントサイズを調整します。タイトルや軸ラベルごとに異なるサイズを設定できます。
6. `plt.grid(True, linestyle='--', alpha=0.7)`でグリッド線を追加します。`linestyle='--'`で破線に、`alpha=0.7`で透明度を設定しています。

## 複数の線を1つのグラフに表示

複数の都市のデータを1つのグラフに表示してみましょう。


```python
# 複数都市のデータ作成
data_multi = {
    '月': list(range(1, 13)),
    '東京': [100, 120, 150, 200, 250, 220, 300, 350, 280, 220, 180, 150],
    '大阪': [80, 90, 110, 150, 180, 160, 220, 260, 210, 170, 140, 120],
    '京都': [60, 70, 90, 120, 150, 130, 180, 210, 170, 140, 110, 90]
}

df_multi = pd.DataFrame(data_multi)
df_multi.set_index('月', inplace=True)

# 複数の折れ線グラフの作成
df_multi.plot(kind='line', figsize=(12, 6), marker='o')
plt.title('主要都市の月別観光客数（2023年）', fontsize=16)
plt.ylabel('観光客数（万人）', fontsize=12)
plt.xlabel('月', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

このコードの新しいポイントと詳細説明：
1. 複数の都市のデータを含む新しいデータフレーム`df_multi`を作成します。
2. `df_multi.plot()`で全ての列（都市）のデータを一度にプロットします。各都市が異なる色の線で表示されます。
3. `plt.legend()`で凡例を追加します。これにより、どの線がどの都市を表しているか分かります。
4. `fontsize=10`で凡例のフォントサイズを設定しています。

問題1
あなたは、過去6ヶ月間の毎月の貯金額を記録しました。このデータを使って、貯金額の推移を表す折れ線グラフを作成してください。

データ：
月 = [1, 2, 3, 4, 5, 6]
貯金額 = [30000, 35000, 28000, 40000, 38000, 45000]

以下の要件を満たす折れ線グラフを作成してください：
1. グラフのタイトルを「過去6ヶ月間の貯金額推移」とする
2. x軸のラベルを「月」、y軸のラベルを「貯金額（円）」とする
3. 線の色を青色にし、データポイントに丸いマーカーを追加する
4. グリッド線を追加する

解答例(Pythonコード):


```python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの準備
data = {
    '月': [1, 2, 3, 4, 5, 6],
    '貯金額': [30000, 35000, 28000, 40000, 38000, 45000]
}

df = pd.DataFrame(data)
df.set_index('月', inplace=True)

# データの確認
print(df)

# 折れ線グラフの作成
df.plot(kind='line', figsize=(10, 6), color='blue', marker='o')
plt.title('過去6ヶ月間の貯金額推移', fontsize=16)
plt.xlabel('月', fontsize=12)
plt.ylabel('貯金額（円）', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

解説:
1. 必要なライブラリ（pandas と matplotlib.pyplot）をインポートします。
2. 日本語フォントを設定して、グラフ上の文字が正しく表示されるようにします。
3. 提供されたデータを辞書形式で定義し、それをpandasのDataFrameに変換します。
4. `set_index('月', inplace=True)` で「月」列をインデックスに設定します。
5. `print(df)` でデータを確認します。
6. `df.plot()` 関数を使って折れ線グラフを作成します。`kind='line'` で折れ線グラフを指定し、`color='blue'` で線の色を青に、`marker='o'` で丸いマーカーを追加します。
7. `plt.title()`, `plt.xlabel()`, `plt.ylabel()` でグラフのタイトルと軸ラベルを設定します。
8. `plt.grid()` でグリッド線を追加します。
9. `plt.show()` でグラフを表示します。

このグラフを見ることで、過去6ヶ月間の貯金額の変動を視覚的に理解することができます。


問題2
あなたとあなたの友人は、1週間毎日の歩数を記録しました。このデータを使って、二人の歩数の推移を比較する折れ線グラフを作成してください。

データ：
日付 = [1, 2, 3, 4, 5, 6, 7]
あなたの歩数 = [8000, 7500, 9000, 8500, 10000, 9500, 7000]
友人の歩数 = [7000, 8000, 8500, 9000, 8000, 9500, 8000]

以下の要件を満たす折れ線グラフを作成してください：
1. グラフのタイトルを「1週間の歩数比較」とする
2. x軸のラベルを「日付」、y軸のラベルを「歩数」とする
3. あなたの歩数を赤色、友人の歩数を青色で表示する
4. 凡例を追加する
5. グリッド線を追加する

解答例(Pythonコード):


```python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの準備
data = {
    '日付': [1, 2, 3, 4, 5, 6, 7],
    'あなた': [8000, 7500, 9000, 8500, 10000, 9500, 7000],
    '友人': [7000, 8000, 8500, 9000, 8000, 9500, 8000]
}

df = pd.DataFrame(data)
df.set_index('日付', inplace=True)

# データの確認
print(df)

# 折れ線グラフの作成
df.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('1週間の歩数比較', fontsize=16)
plt.xlabel('日付', fontsize=12)
plt.ylabel('歩数', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```


解説:
1. 必要なライブラリ（pandas と matplotlib.pyplot）をインポートします。
2. 日本語フォントを設定して、グラフ上の文字が正しく表示されるようにします。
3. 提供されたデータを辞書形式で定義し、それをpandasのDataFrameに変換します。
4. `set_index('日付', inplace=True)` で「日付」列をインデックスに設定します。
5. `print(df)` でデータを確認します。
6. `df.plot()` 関数を使って折れ線グラフを作成します。`kind='line'` で折れ線グラフを指定し、`marker='o'` でマーカーを追加します。
7. `plt.title()`, `plt.xlabel()`, `plt.ylabel()` でグラフのタイトルと軸ラベルを設定します。
8. `plt.legend()` で凡例を追加します。
9. `plt.grid()` でグリッド線を追加します。
10. `plt.show()` でグラフを表示します。

このグラフを見ることで、あなたと友人の1週間の歩数の推移を比較し、視覚的に理解することができます。二人の歩数の傾向や違いを簡単に把握できます。
