
# Matplotlibで日本の観光データを可視化しよう：エリアプロットの作成

## はじめに

こんにちは！このページでは、Pythonのデータ可視化ライブラリMatplotlibを使って、エリアプロットを作成する方法を学びます。エリアプロットは、時間の経過とともに変化する複数の要素の割合や量を表現するのに適したグラフです。

今回は、日本の主要観光地の年間観光客数の推移を例に、エリアプロットの作成方法を説明します。初心者の方にも分かりやすいよう、段階的に進めていきますので、ぜひ最後までお付き合いください。

## エリアプロットとは？

エリアプロットは、線グラフの下の部分を色で塗りつぶしたグラフです。複数の要素の推移を同時に表現でき、それぞれの要素が全体に占める割合も視覚的に理解しやすいのが特徴です。

具体的な特徴は以下の通りです：

1. 時系列データの可視化に適している
2. 複数の要素の推移を一つのグラフで表現できる
3. 積み上げ式にすることで、全体に対する各要素の割合を表現できる
4. 色の濃淡や透明度を調整することで、データの重なりを表現できる

観光客数の推移や、売上の内訳の変化など、時系列データの可視化に適しています。

## データの準備

まず、分析に使用するデータを準備します。今回は、日本の主要観光地の年間観光客数のデータを使用します。


````python
import pandas as pd
import matplotlib.pyplot as plt

# データの作成
data = {
    '年': range(2015, 2022),
    '東京': [5000, 5200, 5500, 5800, 6000, 3000, 3500],
    '京都': [2000, 2100, 2300, 2500, 2700, 1000, 1200],
    '大阪': [2500, 2700, 3000, 3200, 3400, 1500, 1800],
    '北海道': [1800, 2000, 2200, 2400, 2600, 1200, 1400],
    '沖縄': [1000, 1200, 1400, 1600, 1800, 800, 1000]
}

# DataFrameの作成
df = pd.DataFrame(data)
df.set_index('年', inplace=True)

print(df)
````

このコードでは以下の操作を行っています：

1. `pandas`と`matplotlib.pyplot`をインポートします。
2. 観光客数のデータを辞書形式で作成します。
3. `pd.DataFrame()`を使って、辞書からデータフレームを作成します。
4. `set_index('年', inplace=True)`で、'年'列をインデックスに設定します。これにより、年がグラフのx軸になります。
5. `print(df)`でデータフレームの内容を確認します。

データフレームを使うことで、データの操作や可視化が容易になります。

## エリアプロットの作成

次に、Matplotlibを使ってエリアプロットを作成します。


````python
# 日本語フォントの設定
＃plt.rcParams['font.family'] = 'Yu Gothic'  # Windows向け
plt.rcParams['font.family'] = 'AppleGothic'  # Mac向け

# エリアプロットの作成
ax = df.plot(kind='area', figsize=(12, 6), stacked=True, alpha=0.7)

# グラフのカスタマイズ
plt.title('主要観光地の年間観光客数推移', fontsize=16)
plt.xlabel('年', fontsize=12)
plt.ylabel('観光客数（万人）', fontsize=12)
plt.legend(title='観光地', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

# y軸の目盛りを調整
plt.yticks([i*1000 for i in range(17)], [str(i*100) for i in range(17)])

# グラフの表示
plt.tight_layout()
plt.show()
````

このコードの各部分について詳しく説明します：

1. 日本語フォントの設定：
   ```python
   plt.rcParams['font.family'] = 'MS Gothic'  # Windows向け
   # plt.rcParams['font.family'] = 'AppleGothic'  # Mac向け
   ```
   日本語を正しく表示するために、フォントを設定します。OSによって適切なフォントが異なるので注意が必要です。

2. エリアプロットの作成：
   ```python
   ax = df.plot(kind='area', figsize=(12, 6), stacked=True, alpha=0.7)
   ```
   - `kind='area'`でエリアプロットを指定します。
   - `figsize=(12, 6)`でグラフのサイズを設定します。
   - `stacked=True`で積み上げ式のエリアプロットを作成します。
   - `alpha=0.7`で透明度を設定し、重なり合う部分を見やすくします。

3. グラフのカスタマイズ：
   ```python
   plt.title('主要観光地の年間観光客数推移', fontsize=16)
   plt.xlabel('年', fontsize=12)
   plt.ylabel('観光客数（万人）', fontsize=12)
   plt.legend(title='観光地', bbox_to_anchor=(1.05, 1), loc='upper left')
   plt.grid(True, alpha=0.3)
   ```
   タイトル、x軸とy軸のラベル、凡例、グリッド線を設定します。

4. y軸の目盛りの調整：
   ```python
   plt.yticks([i*1000 for i in range(17)], [str(i*100) for i in range(17)])
   ```
   y軸の目盛りを万人単位に調整します。

5. グラフの表示：
   ```python
   plt.tight_layout()
   plt.show()
   ```
   `tight_layout()`でレイアウトを自動調整し、`show()`でグラフを表示します。

## データの分析

エリアプロットを作成したら、データの傾向を分析してみましょう。


# 年ごとの総観光客数
total_tourists = df.sum(axis=1)
print("年ごとの総観光客数（万人）:")
print(total_tourists)

# 観光地ごとの最大観光客数とその年
max_tourists = df.idxmax()
print("\n観光地ごとの最大観光客数とその年:")
for place, year in max_tourists.items():
    print(f"{place}: {df.loc[year, place]}万人 ({year}年)")

# 2019年と2021年の観光客数の比較（コロナの影響）
change_rate = (df.loc[2021] - df.loc[2019]) / df.loc[2019] * 100
print("\n2019年から2021年の観光客数の変化率（%）:")
print(change_rate)


このコードでは、以下の分析を行っています：

1. 年ごとの総観光客数の計算：
   ```python
   total_tourists = df.sum(axis=1)
   ```
   `axis=1`で行方向（年ごと）の合計を計算します。

2. 各観光地の最大観光客数とその年の特定：
   ```python
   max_tourists = df.idxmax()
   ```
   `idxmax()`で各列（観光地）の最大値のインデックス（年）を取得します。

3. 2019年と2021年の観光客数の比較：
   ```python
   change_rate = (df.loc[2021] - df.loc[2019]) / df.loc[2019] * 100
   ```
   2019年と2021年の差分を計算し、変化率を求めます。

これらの分析結果を確認することで、エリアプロットから読み取れる情報をさらに詳しく理解することができます。
