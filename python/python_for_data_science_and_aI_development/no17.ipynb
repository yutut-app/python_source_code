# Pythonで折れ線グラフを作ろう！日本の観光データ可視化

## はじめに

こんにちは！このページでは、Pythonを使って折れ線グラフ（"Line Plot"）を作成する方法を学びます。折れ線グラフは、時間の経過に伴うデータの変化を可視化するのに適しています。ここでは、日本の観光データを例に、折れ線グラフの作り方を step by step で説明していきます。

## 折れ線グラフとは

折れ線グラフは、データポイントを直線で結んだグラフです。以下の特徴があります：

1. 時系列データの傾向を把握するのに適している
2. 複数のデータセットを比較しやすい
3. 連続的な変化を表現できる
4. データの急激な変化や異常値を見つけやすい

観光客数の推移や気温の変化など、時間とともに変化するデータの可視化によく使われます。

## データの準備

まずは、分析に使うデータを準備しましょう。ここでは、東京の月別観光客数のデータを使用します。


```python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの作成
data = {
    '月': list(range(1, 13)),
    '観光客数': [100, 120, 150, 200, 250, 220, 300, 350, 280, 220, 180, 150]
}

df = pd.DataFrame(data)
df.set_index('月', inplace=True)

print(df.head())
```

このコードでは、以下の操作を行っています：
1. `import pandas as pd`と`import matplotlib.pyplot as plt`で必要なライブラリをインポートします。pandasはデータ操作用、matplotlibはグラフ描画用です。
2. `plt.rcParams['font.family']`で日本語フォントを設定します。これにより、グラフ上の日本語が正しく表示されます。
3. `data`辞書で月別観光客数のデータを作成します。実際のデータ収集を模倣しています。
4. `pd.DataFrame(data)`でデータフレームを作成します。これにより、データを表形式で扱えるようになります。
5. `set_index('月', inplace=True)`で'月'列をインデックスに設定します。これにより、月をX軸に使用しやすくなります。
6. `print(df.head())`でデータフレームの最初の5行を表示し、データが正しく読み込まれているか確認します。

## 基本的な折れ線グラフの作成

次に、基本的な折れ線グラフを作成します。


```python
# 基本的な折れ線グラフの作成
df.plot(kind='line', figsize=(10, 6))
plt.title('東京の月別観光客数')
plt.ylabel('観光客数（万人）')
plt.xlabel('月')
plt.show()
```



このコードの詳細説明：
1. `df.plot(kind='line')`で線グラフを指定します。`kind='line'`パラメータにより、折れ線グラフが選択されます。
2. `figsize=(10, 6)`でグラフのサイズを設定します。単位はインチです。
3. `plt.title()`でグラフのタイトルを設定します。
4. `plt.ylabel()`と`plt.xlabel()`で、それぞれY軸とX軸のラベルを設定します。
5. `plt.show()`でグラフを表示します。これがないと、グラフが画面に表示されません。

## グラフのカスタマイズ

グラフをより見やすくカスタマイズしましょう。


```python
# グラフのカスタマイズ
df.plot(kind='line', figsize=(10, 6), color='red', marker='o', linewidth=2, markersize=8)
plt.title('東京の月別観光客数（2023年）', fontsize=16)
plt.ylabel('観光客数（万人）', fontsize=12)
plt.xlabel('月', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

このコードでの追加点と詳細説明：
1. `color='red'`で線の色を赤に設定します。他の色名やRGB値も使用可能です。
2. `marker='o'`でデータポイントにマーカーを追加します。'o'は円形のマーカーを意味します。
3. `linewidth=2`で線の太さを設定します。数値が大きいほど太くなります。
4. `markersize=8`でマーカーのサイズを設定します。数値が大きいほど大きくなります。
5. `fontsize`パラメータでフォントサイズを調整します。タイトルや軸ラベルごとに異なるサイズを設定できます。
6. `plt.grid(True, linestyle='--', alpha=0.7)`でグリッド線を追加します。`linestyle='--'`で破線に、`alpha=0.7`で透明度を設定しています。

## 複数の線を1つのグラフに表示

複数の都市のデータを1つのグラフに表示してみましょう。


```python
# 複数都市のデータ作成
data_multi = {
    '月': list(range(1, 13)),
    '東京': [100, 120, 150, 200, 250, 220, 300, 350, 280, 220, 180, 150],
    '大阪': [80, 90, 110, 150, 180, 160, 220, 260, 210, 170, 140, 120],
    '京都': [60, 70, 90, 120, 150, 130, 180, 210, 170, 140, 110, 90]
}

df_multi = pd.DataFrame(data_multi)
df_multi.set_index('月', inplace=True)

# 複数の折れ線グラフの作成
df_multi.plot(kind='line', figsize=(12, 6), marker='o')
plt.title('主要都市の月別観光客数（2023年）', fontsize=16)
plt.ylabel('観光客数（万人）', fontsize=12)
plt.xlabel('月', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

このコードの新しいポイントと詳細説明：
1. 複数の都市のデータを含む新しいデータフレーム`df_multi`を作成します。
2. `df_multi.plot()`で全ての列（都市）のデータを一度にプロットします。各都市が異なる色の線で表示されます。
3. `plt.legend()`で凡例を追加します。これにより、どの線がどの都市を表しているか分かります。
4. `fontsize=10`で凡例のフォントサイズを設定しています。

## まとめ

このページでは、Pythonを使って折れ線グラフを作成する方法を学びました。折れ線グラフは、時系列データの傾向を視覚化するのに適しています。

主なポイント：
1. `pandas`と`matplotlib`を使用して折れ線グラフを作成できる

3. 色、マーカー、線の太さなどをカスタマイズできる

- 年間の観光客数の推移を可視化し、ピークシーズンを特定する
- 季節ごとの観光トレンドを分析し、イベントの企画に活用する
- 複数の観光地の人気度を比較し、マーケティング戦略を立てる

これらの応用例を参考に、ぜひ自分の興味のある観光データで折れ線グラフを作成してみてください。データの可視化を通じて、新たな洞察が得られるかもしれません。

問題1
あなたは、過去1年間の毎月の読書時間を記録しました。このデータを使って、折れ線グラフを作成し、読書時間の推移を可視化してください。

1. データフレームを作成し、最初の5行を表示してください。
2. 基本的な折れ線グラフを作成し、適切なタイトルと軸ラベルを設定してください。
3. グラフをカスタマイズし、マーカー、線の色、グリッド線を追加してください。

解答例(Pythonコード):


```python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの作成
data = {
    '月': list(range(1, 13)),
    '読書時間': [10, 12, 15, 20, 25, 30, 35, 40, 30, 25, 20, 15]
}

# 1. DataFrameの作成と表示
df = pd.DataFrame(data)
df.set_index('月', inplace=True)
print("読書時間データ（最初の5行）:")
print(df.head())

# 2. 基本的な折れ線グラフの作成
plt.figure(figsize=(10, 6))
df.plot(kind='line')
plt.title('月別読書時間の推移')
plt.ylabel('読書時間（時間）')
plt.xlabel('月')
plt.show()

# 3. カスタマイズされた折れ線グラフの作成
plt.figure(figsize=(10, 6))
df.plot(kind='line', color='green', marker='^', linewidth=2, markersize=8)
plt.title('月別読書時間の推移（カスタマイズ版）', fontsize=16)
plt.ylabel('読書時間（時間）', fontsize=12)
plt.xlabel('月', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(df.index)  # x軸の目盛りを整数に設定
plt.show()
```

解説:
このコードでは、月別の読書時間データを使って折れ線グラフを作成しています。

1. `pd.DataFrame()`でデータフレームを作成し、`set_index('月', inplace=True)`で月を行インデックスに設定しています。

2. 基本的な折れ線グラフでは、`df.plot(kind='line')`を使用しています。`plt.title()`, `plt.ylabel()`, `plt.xlabel()`でグラフのタイトルと軸ラベルを設定しています。

3. カスタマイズされたグラフでは、以下の追加設定を行っています：
   - `color='green'`で線の色を緑に設定
   - `marker='o'`で各データポイントに丸いマーカーを追加
   - `linewidth=2`で線の太さを設定
   - `markersize=8`でマーカーのサイズを設定
   - `plt.grid()`でグリッド線を追加
   - `plt.xticks(df.index)`でx軸の目盛りを整数（月）に設定

これらの設定により、読書時間の推移がより見やすく視覚化されています。夏季（7-8月）に読書時間が増加し、冬季に減少する傾向が見て取れます。

問題2（改良版）
あなたは、3人の友人と一緒に過去6ヶ月間、毎月のカラオケでの歌唱曲数を記録しました。このデータを使って、4人分の折れ線グラフを1つのグラフ上に作成し、カラオケでの歌唱曲数の推移を比較してください。

1. データフレームを作成し、最初の3行を表示してください。
2. 4人分のデータを1つのグラフ上に折れ線グラフとして表示し、適切なタイトル、軸ラベル、凡例を設定してください。
3. グラフをカスタマイズし、マーカー、グリッド線を追加し、各線を識別しやすくしてください。

解答例(Pythonコード):


```python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの作成
data = {
    '月': list(range(1, 7)),
    'あなた': [15, 18, 22, 25, 20, 17],
    '友人A': [10, 14, 18, 20, 16, 12],
    '友人B': [20, 23, 28, 26, 22, 19],
    '友人C': [12, 15, 19, 21, 18, 14]
}

# 1. DataFrameの作成と表示
df = pd.DataFrame(data)
df.set_index('月', inplace=True)
print("カラオケ歌唱曲数データ（最初の3行）:")
print(df.head(3))

# 2. および 3. 複数の折れ線グラフの作成とカスタマイズ
plt.figure(figsize=(12, 6))
for column in df.columns:
    plt.plot(df.index, df[column], marker='s', label=column, linewidth=2, markersize=8)

plt.title('月別カラオケ歌唱曲数の推移', fontsize=16)
plt.ylabel('歌唱曲数', fontsize=12)
plt.xlabel('月', fontsize=12)
plt.legend(title='メンバー', fontsize=10, title_fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(df.index)  # x軸の目盛りを整数に設定
plt.ylim(0, max(df.max()) + 5)  # y軸の範囲を0から最大値+5に設定
plt.show()
```

解説:
このコードでは、4人分のカラオケ歌唱曲数データを使って複数の折れ線グラフを1つのグラフ上に作成しています。以前の問題と比べて、以下の点が改善されています：

1. データの範囲を広げ、各人の歌唱曲数に差をつけることで、線が重なりにくくなりました。

2. `for` ループを使用して各メンバーのデータを個別にプロットしています。これにより、各線のスタイルをより細かく制御できます：
   - `marker='o'`で各データポイントにマーカーを追加
   - `linewidth=2`で線の太さを設定
   - `markersize=8`でマーカーのサイズを設定

3. `plt.ylim(0, max(df.max()) + 5)`でy軸の範囲を0から最大値+5に設定しています。これにより、グラフの下部に余白ができ、データの変化がより明確になります。

4. その他の設定（タイトル、軸ラベル、凡例、グリッド線、x軸の目盛り）は前回と同様です。

このグラフにより、4人のカラオケでの歌唱曲数の推移を明確に比較できます。例えば、友人Bが常に最も多くの曲を歌っていることや、全体的に3-4月頃にピークがあることなどが視覚的に理解しやすくなっています。また、各メンバーの歌唱曲数の差や変化の傾向も明確に見て取れるようになりました。
