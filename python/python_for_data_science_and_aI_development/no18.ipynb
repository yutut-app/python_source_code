# Pythonで面グラフ（エリアプロット）を作ろう！

## はじめに

こんにちは！このページでは、Pythonを使って面グラフ（エリアプロット）を作成する方法を学びます。面グラフは、時間経過による変化や複数の要素の比較を視覚的に表現するのに適しています。日本の観光データを例に、面グラフの作り方を step by step で説明していきます。

## 面グラフとは

面グラフ（"Area Plot"）は、線グラフの下の部分を塗りつぶしたようなグラフです。複数の要素の推移や割合を同時に表現できるため、観光客数の推移などを分かりやすく示すのに適しています。

面グラフの特徴：
1. 時系列データの可視化に適している
2. 複数の要素の比較や構成比を表現できる
3. 積み上げ式と非積み上げ式がある
4. データの累積的な性質を強調できる

## データの準備

まず、分析に使うデータを準備します。ここでは、日本の主要観光地の年間観光客数データを使用します。


```python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの作成
data = {
    '年': range(2010, 2021),
    '東京': [800, 850, 900, 950, 1000, 1100, 1200, 1300, 1400, 1000, 500],
    '大阪': [600, 650, 700, 750, 800, 850, 900, 950, 1000, 700, 400],
    '京都': [500, 550, 600, 650, 700, 750, 800, 850, 900, 600, 300],
    '北海道': [400, 450, 500, 550, 600, 650, 700, 750, 800, 500, 250]
}

df = pd.DataFrame(data)
df.set_index('年', inplace=True)

print(df.head())
```

このコードの詳細説明：
1. 必要なライブラリ（pandas と matplotlib）をインポートします。
2. 日本語フォントを設定します。これにより、グラフ上の日本語が正しく表示されます。
3. 辞書形式でデータを作成します。これは、実際のデータ収集を模倣しています。
4. `pd.DataFrame()` でデータフレームを作成します。これにより、データを表形式で扱えるようになります。
5. `set_index('年', inplace=True)` で、'年'列をインデックスに設定します。これにより、年をX軸に使用しやすくなります。
6. `print(df.head())` で、データフレームの最初の5行を表示し、正しくデータが読み込まれているか確認します。

## 基本的な面グラフの作成

次に、基本的な面グラフを作成します。


```python
# 基本的な面グラフの作成
df.plot(kind='area', figsize=(12, 6), alpha=1.0)
plt.title('日本の主要観光地の年間観光客数推移')
plt.ylabel('観光客数（万人）')
plt.xlabel('年')
plt.show()
```

このコードの詳細説明：
1. `df.plot()` メソッドを使用して、データフレームからグラフを作成します。
2. `kind='area'` で、グラフの種類を面グラフに指定します。
3. `figsize=(12, 6)` でグラフのサイズを設定します。
4. `plt.title()`, `plt.ylabel()`, `plt.xlabel()` でグラフのタイトルと軸ラベルを設定します。
5. `plt.show()` でグラフを表示します。

## 積み上げなしの面グラフ

デフォルトでは面グラフは積み上げ式になりますが、`stacked=False` を指定することで、積み上げなしの面グラフを作成できます。


```python
# 積み上げなしの面グラフ
df.plot(kind='area', stacked=False, figsize=(12, 6), alpha=1.0)
plt.title('日本の主要観光地の年間観光客数推移（積み上げなし）')
plt.ylabel('観光客数（万人）')
plt.xlabel('年')
plt.show()
```

このコードの詳細説明：
1. `stacked=False` を追加し、積み上げなしの面グラフを指定します。
2. その他のパラメータは基本的な面グラフと同じです。
3. 積み上げなしの面グラフは、各要素の絶対的な値の比較に適しています。

## グラフのカスタマイズ

matplotlib の機能を使って、グラフをさらにカスタマイズできます。


```python
# グラフのカスタマイズ
ax = df.plot(kind='area', stacked=True, figsize=(12, 6), alpha=0.7)
ax.set_title('日本の主要観光地の年間観光客数推移（カスタマイズ版）', fontsize=16)
ax.set_ylabel('観光客数（万人）', fontsize=12)
ax.set_xlabel('年', fontsize=12)
ax.legend(loc='upper left', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

このコードの詳細説明：
1. `ax = df.plot()` で、グラフオブジェクト（Axes）を取得します。
2. `ax.set_title()`, `ax.set_ylabel()`, `ax.set_xlabel()` で、タイトルと軸ラベルをカスタマイズします。
3. `fontsize` パラメータでフォントサイズを調整します。
4. `alpha=0.7` で透明度を設定し、重なり合う部分も見やすくします。
5. `ax.legend()` で凡例の位置とフォントサイズを設定します。
6. `plt.grid()` でグリッド線を追加し、`linestyle` と `alpha` でスタイルを調整します。

## まとめ

このページでは、Pythonを使って面グラフ（エリアプロット）を作成する方法を学びました。面グラフは、時系列データの推移や複数要素の比較を視覚的に表現するのに適しています。

主なポイントは以下の通りです：
1. pandas と matplotlib を使用して面グラフを作成できる
2. `df.plot(kind='area')` で基本的な面グラフを作成
3. `stacked=False` で積み上げなしの面グラフを作成可能
4. `alpha` パラメータで透明度を調整できる
5. matplotlib の機能を使ってグラフを詳細にカスタマイズできる

面グラフは、観光データの分析や可視化に非常に有効です。例えば、複数の観光地の観光客数の推移を比較したり、ある観光地の季節ごとの観光客数の割合を表現したりするのに適しています。

応用例：
1. 年間の観光収入の内訳を地域別に表示
2. 宿泊施設の種類別の利用率の推移を可視化
3. 観光スポットの人気度の変化を時系列で比較

ぜひ、自分の興味のあるデータで面グラフを作成してみてください。データの特性に応じて、積み上げ式や非積み上げ式を選択し、適切な透明度や色使いでグラフをカスタマイズすることで、より効果的なデータの可視化が可能になります。

問題1
あなたは、過去5年間の自分の月別支出を記録しました。主な支出項目である「食費」「住居費」「交通費」について、面グラフを作成して支出の推移を可視化してください。

データ：
年 = [2019, 2020, 2021, 2022, 2023]
食費 = [50000, 48000, 52000, 55000, 58000]
住居費 = [80000, 82000, 83000, 85000, 87000]
交通費 = [15000, 12000, 13000, 14000, 16000]

以下の要件を満たす面グラフを作成してください：
1. グラフのタイトルを「過去5年間の月別支出推移」とする
2. x軸のラベルを「年」、y軸のラベルを「支出（円）」とする
3. 積み上げ式の面グラフを作成する
4. 凡例を左上に配置する
5. グリッド線を追加する
6. グラフの透明度を0.7に設定する

解答例(Pythonコード):


```python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの準備
data = {
    '年': [2019, 2020, 2021, 2022, 2023],
    '食費': [50000, 48000, 52000, 55000, 58000],
    '住居費': [80000, 82000, 83000, 85000, 87000],
    '交通費': [15000, 12000, 13000, 14000, 16000]
}

df = pd.DataFrame(data)
df.set_index('年', inplace=True)

# データの確認
print(df)

# 面グラフの作成
ax = df.plot(kind='area', stacked=True, figsize=(10, 6), alpha=0.7)
ax.set_title('過去5年間の月別支出推移', fontsize=16)
ax.set_xlabel('年', fontsize=12)
ax.set_ylabel('支出（円）', fontsize=12)
ax.legend(loc='upper left', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# x軸を整数で表示
ax.set_xticks(df.index)
ax.set_xticklabels(df.index.astype(int))

plt.show()
```

解説:
1. 必要なライブラリ（pandas と matplotlib.pyplot）をインポートします。
2. 日本語フォントを設定して、グラフ上の文字が正しく表示されるようにします。
3. 提供されたデータを辞書形式で定義し、それをpandasのDataFrameに変換します。
4. `set_index('年', inplace=True)` で「年」列をインデックスに設定します。
5. `print(df)` でデータを確認します。
6. `df.plot()` 関数を使って面グラフを作成します。`kind='area'` で面グラフを指定し、`stacked=True` で積み上げ式に、`alpha=0.7` で透明度を設定します。
7. `ax.set_title()`, `ax.set_xlabel()`, `ax.set_ylabel()` でグラフのタイトルと軸ラベルを設定します。
8. `ax.legend()` で凡例を左上に配置します。
9. `plt.grid()` でグリッド線を追加します。
10. `ax.set_xticks()` と `ax.set_xticklabels()` でx軸を整数で表示するように設定します。
11. `plt.show()` でグラフを表示します。

このグラフを見ることで、過去5年間の月別支出の推移と各項目の割合を視覚的に理解することができます。


問題2
あなたは、地元の図書館で過去5年間の分野別貸出冊数データを入手しました。主要な3分野（「文学」「科学」「歴史」）について、面グラフを作成して貸出冊数の推移を可視化してください。

データ：
年度 = [2019, 2020, 2021, 2022, 2023]
文学 = [5000, 5200, 4800, 4300, 4100]
科学 = [3000, 3500, 4000, 4200, 4500]
歴史 = [2000, 2100, 2300, 2500, 2700]

以下の要件を満たす面グラフを作成してください：
1. グラフのタイトルを「図書館の分野別貸出冊数推移」とする
2. x軸のラベルを「年度」、y軸のラベルを「貸出冊数」とする
3. 積み上げなしの面グラフを作成する
4. 凡例を右上に配置する
5. グリッド線を追加する
6. グラフの透明度を0.5に設定する

解答例(Pythonコード):


```python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの準備
data = {
    '年度': [2019, 2020, 2021, 2022, 2023],
    '文学': [5000, 5200, 4800, 4300, 4100],
    '科学': [3000, 3500, 4000, 4200, 4500],
    '歴史': [2000, 2100, 2300, 2500, 2700]
}

df = pd.DataFrame(data)
df.set_index('年度', inplace=True)

# データの確認
print(df)

# 面グラフの作成
ax = df.plot(kind='area', stacked=False, figsize=(10, 6), alpha=0.5)
ax.set_title('図書館の分野別貸出冊数推移', fontsize=16)
ax.set_xlabel('年度', fontsize=12)
ax.set_ylabel('貸出冊数', fontsize=12)
ax.legend(loc='upper right', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# x軸を整数で表示
ax.set_xticks(df.index)
ax.set_xticklabels(df.index.astype(int))

plt.show()
```

1. 必要なライブラリ（pandas と matplotlib.pyplot）をインポートします。
2. 日本語フォントを設定して、グラフ上の文字が正しく表示されるようにします。

8. `ax.legend()` で凡例を右上に配置します。
9. `plt.grid()` でグリッド線を追加します。
10. `ax.set_xticks()` と `ax.set_xticklabels()` でx軸を整数で表示するように設定します。

このグラフを見ることで、過去5年間の図書館の分野別貸出冊数の推移を視覚的に理解することができます。積み上げなしの面グラフを使用することで、各分野の貸出冊数の絶対的な変化を比較しやすくなっています。
