# Pythonで面グラフ（エリアプロット）を作ろう！

## はじめに

こんにちは！このページでは、Pythonを使って面グラフ（エリアプロット）を作成する方法を学びます。面グラフは、時間経過による変化や複数の要素の比較を視覚的に表現するのに適しています。日本の観光データを例に、面グラフの作り方を step by step で説明していきます。

## 面グラフとは

面グラフ（"Area Plot"）は、線グラフの下の部分を塗りつぶしたようなグラフです。複数の要素の推移や割合を同時に表現できるため、観光客数の推移などを分かりやすく示すのに適しています。

面グラフの特徴：
1. 時系列データの可視化に適している
2. 複数の要素の比較や構成比を表現できる
3. 積み上げ式と非積み上げ式がある
4. データの累積的な性質を強調できる

## データの準備

まず、分析に使うデータを準備します。ここでは、日本の主要観光地の年間観光客数データを使用します。


```python
import pandas as pd
import matplotlib.pyplot as plt

# フォントの設定（日本語表示のため）
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

# データの作成
data = {
    '年': range(2010, 2021),
    '東京': [800, 850, 900, 950, 1000, 1100, 1200, 1300, 1400, 1000, 500],
    '大阪': [600, 650, 700, 750, 800, 850, 900, 950, 1000, 700, 400],
    '京都': [500, 550, 600, 650, 700, 750, 800, 850, 900, 600, 300],
    '北海道': [400, 450, 500, 550, 600, 650, 700, 750, 800, 500, 250]
}

df = pd.DataFrame(data)
df.set_index('年', inplace=True)

print(df.head())
```

このコードの詳細説明：
1. 必要なライブラリ（pandas と matplotlib）をインポートします。
2. 日本語フォントを設定します。これにより、グラフ上の日本語が正しく表示されます。
3. 辞書形式でデータを作成します。これは、実際のデータ収集を模倣しています。
4. `pd.DataFrame()` でデータフレームを作成します。これにより、データを表形式で扱えるようになります。
5. `set_index('年', inplace=True)` で、'年'列をインデックスに設定します。これにより、年をX軸に使用しやすくなります。
6. `print(df.head())` で、データフレームの最初の5行を表示し、正しくデータが読み込まれているか確認します。

## 基本的な面グラフの作成

次に、基本的な面グラフを作成します。


```python
# 基本的な面グラフの作成
df.plot(kind='area', figsize=(12, 6), alpha=1.0)
plt.title('日本の主要観光地の年間観光客数推移')
plt.ylabel('観光客数（万人）')
plt.xlabel('年')
plt.show()
```

このコードの詳細説明：
1. `df.plot()` メソッドを使用して、データフレームからグラフを作成します。
2. `kind='area'` で、グラフの種類を面グラフに指定します。
3. `figsize=(12, 6)` でグラフのサイズを設定します。
4. `plt.title()`, `plt.ylabel()`, `plt.xlabel()` でグラフのタイトルと軸ラベルを設定します。
5. `plt.show()` でグラフを表示します。

## 積み上げなしの面グラフ

デフォルトでは面グラフは積み上げ式になりますが、`stacked=False` を指定することで、積み上げなしの面グラフを作成できます。


```python
# 積み上げなしの面グラフ
df.plot(kind='area', stacked=False, figsize=(12, 6), alpha=1.0)
plt.title('日本の主要観光地の年間観光客数推移（積み上げなし）')
plt.ylabel('観光客数（万人）')
plt.xlabel('年')
plt.show()
```

このコードの詳細説明：
1. `stacked=False` を追加し、積み上げなしの面グラフを指定します。
2. その他のパラメータは基本的な面グラフと同じです。
3. 積み上げなしの面グラフは、各要素の絶対的な値の比較に適しています。

## グラフのカスタマイズ

matplotlib の機能を使って、グラフをさらにカスタマイズできます。


```python
# グラフのカスタマイズ
ax = df.plot(kind='area', stacked=True, figsize=(12, 6), alpha=0.7)
ax.set_title('日本の主要観光地の年間観光客数推移（カスタマイズ版）', fontsize=16)
ax.set_ylabel('観光客数（万人）', fontsize=12)
ax.set_xlabel('年', fontsize=12)
ax.legend(loc='upper left', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

このコードの詳細説明：
1. `ax = df.plot()` で、グラフオブジェクト（Axes）を取得します。
2. `ax.set_title()`, `ax.set_ylabel()`, `ax.set_xlabel()` で、タイトルと軸ラベルをカスタマイズします。
3. `fontsize` パラメータでフォントサイズを調整します。
4. `alpha=0.7` で透明度を設定し、重なり合う部分も見やすくします。
5. `ax.legend()` で凡例の位置とフォントサイズを設定します。
6. `plt.grid()` でグリッド線を追加し、`linestyle` と `alpha` でスタイルを調整します。

## まとめ

このページでは、Pythonを使って面グラフ（エリアプロット）を作成する方法を学びました。面グラフは、時系列データの推移や複数要素の比較を視覚的に表現するのに適しています。

主なポイントは以下の通りです：
1. pandas と matplotlib を使用して面グラフを作成できる
2. `df.plot(kind='area')` で基本的な面グラフを作成
3. `stacked=False` で積み上げなしの面グラフを作成可能
4. `alpha` パラメータで透明度を調整できる
5. matplotlib の機能を使ってグラフを詳細にカスタマイズできる

面グラフは、観光データの分析や可視化に非常に有効です。例えば、複数の観光地の観光客数の推移を比較したり、ある観光地の季節ごとの観光客数の割合を表現したりするのに適しています。

応用例：
1. 年間の観光収入の内訳を地域別に表示
2. 宿泊施設の種類別の利用率の推移を可視化
3. 観光スポットの人気度の変化を時系列で比較

ぜひ、自分の興味のあるデータで面グラフを作成してみてください。データの特性に応じて、積み上げ式や非積み上げ式を選択し、適切な透明度や色使いでグラフをカスタマイズすることで、より効果的なデータの可視化が可能になります。

```python
import pandas as pd
import matplotlib.pyplot as plt

# データの作成
data = {
    '年': [2018, 2019, 2020, 2021, 2022],
    '読書': [30, 35, 50, 45, 40],
    '映画鑑賞': [20, 25, 40, 35, 30],
    'スポーツ': [40, 35, 20, 30, 35],
    '料理': [15, 20, 30, 25, 30]
}

# 1. DataFrameの作成と表示
df = pd.DataFrame(data)
df.set_index('年', inplace=True)
print("趣味活動時間データ（最初の3行）:")
print(df.head(3))

# 2. 積み上げ式の面グラフの作成
plt.figure(figsize=(12, 6))
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

ax = df.plot(kind='area', stacked=True, alpha=0.7)
ax.set_title('季節ごとの趣味活動時間の推移', fontsize=16)
ax.set_xlabel('年', fontsize=12)
ax.set_ylabel('活動時間（時間/月）', fontsize=12)
ax.legend(title='趣味', loc='upper left', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(df.index)  # x軸の目盛りを整数に設定
plt.show()

# 3. データの分析
yearly_total = df.sum(axis=1)
max_year = yearly_total.idxmax()
min_year = yearly_total.idxmin()

print("\n各年の合計活動時間:")
print(yearly_total)
print(f"\n最も活動時間が多かった年: {max_year}年 ({yearly_total[max_year]}時間/月)")
print(f"最も活動時間が少なかった年: {min_year}年 ({yearly_total[min_year]}時間/月)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# データの作成
data = {
    '年': [2017, 2018, 2019, 2020, 2021, 2022],
    '食材': [30000, 32000, 33000, 38000, 36000, 35000],
    '外食': [15000, 18000, 20000, 12000, 14000, 16000],
    '惣菜': [10000, 12000, 13000, 16000, 15000, 14000],
    '飲料': [5000, 6000, 7000, 8000, 7500, 7000]
}

# 1. DataFrameの作成と表示
df = pd.DataFrame(data)
df.set_index('年', inplace=True)
print("月ごとの食費データ（最初の3行）:")
print(df.head(3))

# 2. 非積み上げ式の面グラフの作成
plt.figure(figsize=(12, 6))
plt.rcParams['font.family'] = 'Hiragino Sans'  # Macの場合
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windowsの場合

ax = df.plot(kind='area', stacked=False, alpha=0.7)
ax.set_title('月ごとの食費の推移', fontsize=16)
ax.set_xlabel('年', fontsize=12)
ax.set_ylabel('食費（円/月）', fontsize=12)
ax.legend(title='カテゴリ', loc='upper left', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(df.index)  # x軸の目盛りを整数に設定
plt.show()

# 3. データの分析
category_average = df.mean()
max_category = category_average.idxmax()
min_category = category_average.idxmin()

print("\n各食費カテゴリの平均支出額:")
print(category_average)
print(f"\n最も支出が多いカテゴリ: {max_category} ({category_average[max_category]:.0f}円/月)")
print(f"最も支出が少ないカテゴリ: {min_category} ({category_average[min_category]:.0f}円/月)")
```
