```python
# 日本語フォントの設定
# plt.rcParams['font.family'] = 'Yu Gothic'  # Windows向け
plt.rcParams['font.family'] = 'Hiragino Sans'  # Mac向け

# ヒストグラムの作成
plt.figure(figsize=(12, 6))
plt.hist(df['観光客数（万人）'], bins=10, edgecolor='black')

# グラフのカスタマイズ
plt.title('都道府県別年間観光客数の分布', fontsize=16)
plt.xlabel('観光客数（万人）', fontsize=12)
plt.ylabel('都道府県数', fontsize=12)
plt.grid(True, alpha=0.3)

# グラフの表示
plt.tight_layout()
plt.show()
```

```python
# NumPyを使ったヒストグラムの作成
plt.figure(figsize=(12, 6))

# ヒストグラムのデータとビンの計算
counts, bin_edges = np.histogram(df['観光客数（万人）'], bins=10)

# ヒストグラムの描画
plt.hist(df['観光客数（万人）'], bins=bin_edges, edgecolor='black')

# グラフのカスタマイズ
plt.title('都道府県別年間観光客数の分布（改善版）', fontsize=16)
plt.xlabel('観光客数（万人）', fontsize=12)
plt.ylabel('都道府県数', fontsize=12)
plt.grid(True, alpha=0.3)

# x軸の目盛りを設定
plt.xticks(bin_edges)

# グラフの表示
plt.tight_layout()
plt.show()

# ビンの情報を表示
print("ビンの境界値:")
for i, edge in enumerate(bin_edges[:-1]):
    print(f"ビン {i+1}: {edge:.0f} - {bin_edges[i+1]:.0f} 万人")
```

```python
# 色の設定
color_map = {
    '北海道': '#FF9999', '東北': '#FFCC99', '関東': '#FFFF99', '中部': '#99FF99',
    '近畿': '#99FFFF', '中国': '#9999FF', '四国': '#FF99FF', '九州': '#FFCCCC',
    '沖縄': '#CCCCFF'
}

# ヒストグラムの作成
plt.figure(figsize=(14, 7))

for region in color_map.keys():
    region_data = df[df['地方'] == region]['観光客数（万人）']
    plt.hist(region_data, bins=10, alpha=0.7, label=region, color=color_map[region], edgecolor='black')

# グラフのカスタマイズ
plt.title('地方別の都道府県年間観光客数の分布', fontsize=16)
plt.xlabel('観光客数（万人）', fontsize=12)
plt.ylabel('都道府県数', fontsize=12)
plt.legend(title='地方', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

# グラフの表示
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# 正規化されたヒストグラムを描画
plt.hist(df['観光客数2019（万人）'], bins=15, alpha=0.5, label='2019年', density=True)
plt.hist(df['観光客数2020（万人）'], bins=15, alpha=0.5, label='2020年', density=True)

# グラフのカスタマイズ
plt.title('2019年と2020年の都道府県別年間観光客数の正規化された分布比較', fontsize=16)
plt.xlabel('観光客数（万人）', fontsize=12)
plt.ylabel('確率密度', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()

# グラフの表示
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# カーネル密度推定を描画
df['観光客数2019（万人）'].plot.kde(label='2019年')
df['観光客数2020（万人）'].plot.kde(label='2020年')

# グラフのカスタマイズ
plt.title('2019年と2020年の都道府県別年間観光客数の分布（カーネル密度推定）', fontsize=16)
plt.xlabel('観光客数（万人）', fontsize=12)
plt.ylabel('確率密度', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()

# グラフの表示
plt.tight_layout()
plt.show()
```
