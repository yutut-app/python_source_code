```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# データの作成
prefectures = ['北海道', '東京都', '大阪府', '京都府', '沖縄県'] + ['その他の県'] * 42
tourists = np.random.randint(100, 10000, size=47)  # 100万人から1億人の範囲でランダムなデータを生成

# DataFrameの作成
df = pd.DataFrame({'都道府県': prefectures, '観光客数（万人）': tourists})

print(df)
```

```python
# 日本語フォントの設定
#plt.rcParams['font.family'] = 'Yu Gothic'  # Windows向け
plt.rcParams['font.family'] = 'Hiragino Sans'  # Mac向け

# ヒストグラムの作成
plt.figure(figsize=(10, 6))
plt.hist(df['観光客数（万人）'], bins=10, edgecolor='black')

# グラフのカスタマイズ
plt.title('都道府県別年間観光客数の分布', fontsize=16)
plt.xlabel('観光客数（万人）', fontsize=12)
plt.ylabel('都道府県数', fontsize=12)
plt.grid(True, alpha=0.3)

# グラフの表示
plt.tight_layout()
plt.show()
```

```python
# NumPyを使ったヒストグラムの作成
plt.figure(figsize=(10, 6))

# ヒストグラムのデータとビンの計算
counts, bin_edges = np.histogram(df['観光客数（万人）'], bins=10)

# ヒストグラムの描画
plt.hist(df['観光客数（万人）'], bins=bin_edges, edgecolor='black')

# グラフのカスタマイズ
plt.title('都道府県別年間観光客数の分布（改善版）', fontsize=16)
plt.xlabel('観光客数（万人）', fontsize=12)
plt.ylabel('都道府県数', fontsize=12)
plt.grid(True, alpha=0.3)

# x軸の目盛りを設定
plt.xticks(bin_edges)

# グラフの表示
plt.tight_layout()
plt.show()

# ビンの情報を表示
print("ビンの境界値:")
for i, edge in enumerate(bin_edges[:-1]):
    print(f"ビン {i+1}: {edge:.0f} - {bin_edges[i+1]:.0f} 万人")
```

```python
# 基本的な統計量の計算
mean_tourists = df['観光客数（万人）'].mean()
median_tourists = df['観光客数（万人）'].median()
std_tourists = df['観光客数（万人）'].std()

print(f"平均観光客数: {mean_tourists:.2f}万人")
print(f"中央値観光客数: {median_tourists:.2f}万人")
print(f"標準偏差: {std_tourists:.2f}万人")

# 最多と最少の観光客数を持つ都道府県
max_prefecture = df.loc[df['観光客数（万人）'].idxmax()]
min_prefecture = df.loc[df['観光客数（万人）'].idxmin()]

print(f"\n最も観光客数が多い都道府県: {max_prefecture['都道府県']} ({max_prefecture['観光客数（万人）']}万人)")
print(f"最も観光客数が少ない都道府県: {min_prefecture['都道府県']} ({min_prefecture['観光客数（万人）']}万人)")
```
